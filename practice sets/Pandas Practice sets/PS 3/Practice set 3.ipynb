```python
#  12.
Employees = pd.DataFrame({
   "EmpID": [1,2,3],
    "Name": ["A","B","C"],
    "DptID": [10,20,30]
})

Department= pd.DataFrame({
    "DptID": [10,20,30], 
    "DeptName": ["HR","Engineering","Marketing"]
 })
```

```python
# 13.

import pandas as pd

# Left join employees with departments
result = Employees.merge(Department, on="DptID", how="left")
result
```

```python
# 14. 

import pandas as pd

# January data
df_jan = pd.DataFrame({
    "ID": [1, 2, 3],
    "Sales": [100, 200, 150]
})

# February data
df_feb = pd.DataFrame({
    "ID": [4, 5, 6],
    "Sales": [180, 220, 170]
})

print("January DataFrame:")
print(df_jan)

print("\nFebruary DataFrame:")
print(df_feb)
```

```python
combined_df = pd.concat([df_jan, df_feb], ignore_index=True)

print(combined_df)
```

```python
# 16.

import pandas as pd

# Create left_df
left_df = pd.DataFrame({
    "user_id": [1, 2, 3, 4],
    "name": ["Om", "Raj", "Sam", "Amit"]
})

# Create right_df
right_df = pd.DataFrame({
    "id": [1, 2, 4],
    "score": [85, 90, 95]
})

# Left join
result = pd.merge(
    left_df,
    right_df,
    left_on="user_id",
    right_on="id",
    how="left"
)

print(result)
```

```python
# 17.

import pandas as pd

df = pd.DataFrame({
    "Region": ["North", "North", "South", "South", "East", "East"],
    "Product": ["A", "B", "A", "B", "A", "B"],
    "Sales": [100, 150, 200, 250, 300, 350]
})


result = df.groupby(["Region", "Product"])["Sales"].sum().unstack()

print(result)
```

```python
# 18.

import pandas as pd

df = pd.DataFrame({
    "JoinDate": ["2026-01-10", "2026-01-15", "2026-02-05",
                 "2026-02-20", "2026-03-12"]
})

# Convert to datetime
df["JoinDate"] = pd.to_datetime(df["JoinDate"])

# Count rows per month
result = df.groupby(df["JoinDate"].dt.to_period("M")).size()

print(result)
```
