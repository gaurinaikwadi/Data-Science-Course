# Plotting Charts using Matplotlib

### matplotlib.pyplot
matplotlib.pyplot is a module in Matplotlib — it's like a paintbrush for your data. <br>
We usually import it as plt:

```python
import matplotlib.pyplot as plt
```

# plt.show()
plt.show() is used to display the plot. <br>
Without it, in scripts, you might not see the plot window.

```python
plt.plot([1, 2, 3], [4, 5, 6])
```

```python
plt.show()    # used in case of executing a .py file
```

```python
years = [1990, 1992, 1994, 1996, 1998, 2000, 2003, 2005, 2007, 2010]
runs =  [500, 700, 1100, 1500, 1800, 1200, 1700, 1300, 900, 1500]

plt.plot(years, runs)
```

```python
# adding x and y labels
plt.plot(years, runs)            
plt.xlabel("Year")
plt.ylabel("Runs Scored")
plt.title("Sachin Tendulkar's Yearly Runs")      
```

```python
# mutiple lines in one plot

kohli = [0, 0, 500, 800, 1100, 1300, 1500, 1800, 1900, 2100]
sehwag = [0, 300, 800, 1200, 1500, 1700, 1600, 1400, 1000, 0]

plt.plot(years, kohli, label="Virat Kohli")
plt.plot(years, sehwag, label="Virender Sehwag")

plt.xlabel("Year")
plt.ylabel("Runs Scored")
plt.title("Performance Comparison")
plt.legend()
```

```python
# bad practice
plt.plot(years, kohli)
plt.plot(years, sehwag)
plt.legend(["Kohli", "Sehwag"])  # prone to mismatch
```

```python
# better practice

plt.plot(years, kohli, label="Kohli")
plt.plot(years, sehwag, label="Sehwag")
plt.legend()
```

```python
# using format strings

plt.plot(years, kohli, 'ro--', label="Kohli")  # red circles with dashed lines
plt.plot(years, sehwag, 'g^:', label="Sehwag")  # green triangles dotted
plt.legend()
```

###  fmt = '[marker][line][color]'

```python
plt.plot(years, kohli, 'h--c', label="Kohli")  # red circles with dashed lines
plt.plot(years, sehwag, 'D:m', label="Sehwag")  # green triangles dotted
plt.legend()
```

```python
plt.plot(years, kohli, '^k:', label="Kohli")  # red circles with dashed lines
plt.plot(years, sehwag, 'ob-.', label="Sehwag")  # green triangles dotted
plt.legend()
```

```python
plt.plot(years, kohli, color='orange', linestyle='--', label="Kohli")
plt.plot(years, sehwag, color='green', linestyle='-.', label="Sehwag")
plt.plot(years, runs, color='blue', label="Tendulkar")
plt.legend()
```

```python
# line width and layout Tweaks

plt.plot(years, kohli, linewidth=3, label="Kohli")
plt.plot(years, sehwag, linewidth=2, label="Sehwag")
plt.legend()
plt.grid(True)
plt.tight_layout()
```

```python
import numpy as np

for i in range(50):
    plt.plot(np.random.rand(100), linewidth=1)

plt.title("Too Much Data Can Be Confusing!")
plt.grid(True)
plt.tight_layout()
```
