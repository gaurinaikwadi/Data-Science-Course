```python
import json
```

```python
 
# Load the JSON file
def load_data(filename):
    with open(filename, "r") as f:
        data = json.load(f)
    return data
```

```python
data=load_data("data.json")
```

```python
data
```

```python
def display_users(data):
    print("users and thire connections")
    for user in data["users"]:
        print(f"ID:{user['id']} .{user['name']} is Friends with: {user['friends']}  Liked Pages: {user['liked_pages']}")
    print("\nPages:")
    for page in data["pages"]:
        print(f"{page['id']}: {page['name']}")    
```

```python
display_users(data)
```
