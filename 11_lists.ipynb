# List and List Methods
A list in Python is an ordered, mutable collection of elements. It can contain elements of different types.

```python
# creating list

# Empty list
my_list = []

# List with elements
numbers = [1, 2, 3, 4, 5]

# Mixed data types
mixed_list = [1, "Hello", 3.14, True]
```

```python
print(numbers)
```

```python
fruits = ["apple", "banana", "cherry"]
fruits.append("orange")
print(fruits)  

#Sorts the list in ascending order.
fruits.sort()
print(fruits) 

#Extends the list by appending all elements from an iterable.
fruits.extend(["mango", "papaya"])
print(fruits)

#Removes and returns the element at index (last element if index is not provided).
fruits.pop(5)
print(fruits) 

#Reverses the order of the list.
fruits.reverse()
print(fruits) 

```
