```python
a-b
```

```python
a/b
```

**2. Comparison Operators**<br>
Compare values and return True or False.

```python
 a==b
```

```python
a!=b
```

```python
a>b
```

```python
a<b
```

```python
a>=b
```

```python
a<=b
```

**3. Logical Operators**<br>
Used to combine conditional statements.

```python
x = True
y = False
```

```python
x and y
```

```python
x or y
```

```python
not x   #returns true as false and false as true
```

**4. Bitwise Operators**<br>
Perform bit-level operations.

```python
a=5
b=3
```

```python
a & b
```

```python
a | b
```
