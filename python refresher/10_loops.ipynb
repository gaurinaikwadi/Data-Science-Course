# Loops in Python
Python has two main loops: for and while.

## **1. For loop** 
Used to iterate over sequences like lists, tuples, and strings.

```python
fruits = ["gauri", "neha", "shruti"]

for fruit in fruits:
    print(fruit)
```

**Using range()**

```python
for i in range(10):
    print(i)
```

## **2. While loop** 
Runs as long as a condition is True.

```python
count = 0
while count < 9:
    print(count)
    count += 1
```

## 3. Loop Control Statements
break → Exits the loop. <br>
continue → Skips to the next iteration.<br>
pass → Does nothing (used as a placeholder).<br>

```python
for i in range(9):
    if i == 8:
        break  # Stops the loop at 3
    print(i)
```
