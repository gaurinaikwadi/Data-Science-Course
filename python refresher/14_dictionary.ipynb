# Dictionary and Dictionary methods
**Dictionary is:**
- unordered
- mutable

```python
#empty dictionary
empty_dict = {} 

#dictionary with value pairs
student = {
    "name": "Gauri",
     "age": 19,
     "grade": "A+"
}

#using dict construtor
person= dict(name="Gauri", age="20", city="nashik")

print(student)
```

```python
#using key
print(student["name"])

#using get() : avoids keyerror if key dosen't exist
print(student.get("age"))
print(student.get("height","not found"))
```

```python
#adding new key value pair
student["city"]="pune"

#updating an existing value
student["age"]=20

#iterating over a dictionary
for key, value in student.items():
    print(key, ":", value)
```
