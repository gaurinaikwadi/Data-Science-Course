# Data Transformation

## Sorting & Ranking
###  Sort by Values

```python
import pandas as pd
```

```python
df = pd.read_csv("data.csv")
```

```python
df.sort_values("Year")                   # Ascending sort
```

```python
df.sort_values("Year", ascending=False)     # Descending
```

### Reset Index
If you want the index to start from 0 and be sequential, you can reset it using reset_index()

```python
df.reset_index(drop=True, inplace=True)     # Reset the index and drop the old index
```

```python
df
```

```python
df.sort_index()    #sort by index
```

## Ranking

```python
df["rank"] = df["IMDb"].rank()                 # Default: average method
```

```python
df
```

```python
df["Rank"] = df["IMDb"].rank(method="dense")   # 1, 2, 2, 3
```

```python
df
```

### Renaming Columns & Index

```python
df.rename(columns={"oldName": "newName"}, inplace=True)
```

```python
df
```

```python
df.rename(index={0: "row1", 1: "row2"}, inplace=True)
```

```python
df
```

```python
df.columns = ["Actor", "Year","Film", "IMDb","Genre","BoxOffice(INR Crore)","rank","Rank"]
```

```python
df
```

### Changing Column Order
Just pass a new list of column names:

```python
df = df[["Actor", "Year","Film", "IMDb","Genre","BoxOffice(INR Crore)","rank","Rank"]]   # Reorder as desired
```

```python
df
```

```python
cols = ["Actor"] + [col for col in df.columns if col != "Actor"]
df = df[cols]
```

```python
df
```
