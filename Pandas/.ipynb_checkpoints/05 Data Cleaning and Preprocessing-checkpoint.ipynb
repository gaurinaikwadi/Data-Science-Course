# Data Cleaning and Preprocessing

```python
import pandas as pd
```

```python
df = pd.read_csv("Data cleaning and preprocessing.csv")
```

```python
df
```

### check missing data

```python
df.isnull()
```

```python
df.isnull().sum()
```

### drop missing data

```python
df.dropna()
```

```python
df.dropna(axis=1)
```

### fill missing data

```python
df.fillna(0)
```

```python
df.fillna("nothing")
```

```python
df["Marks"].fillna(df["Marks"].mean())
```

```python
df.ffill()        #forward fill
```

```python
df.bfill()          #backward fill
```

### Detecting and Removing Duplicates

```python
df.duplicated()         #return booleans: true for duplicates
```

```python
df.drop_duplicates()    #remove duplicate rows
```

```python
df.duplicated(subset=["Name","Marks"])
```

### String operations with  .str

```python
df["Name"].str.lower()           #convert all names to lowercase
```

```python
df["subject"].str.contains("math", case=False)
```

```python
df["subject"].str.split("@")
```

### Type conversion with  .astype()

```python
df["Marks"]=df["Marks"].astype(float)
```

```python
df
```

```python
df["Marks"]=pd.to_datetime(df["Marks"])
```

```python
df
```

```python
df["category"]=df["category"].astype("category")
```

```python
df
```
