# Data Selection & Filtering
### Selecting Rows & Columns
**Selecting Columns**

```python
import pandas as pd
```

```python
df= pd.read_csv("data.csv")
```

```python
df
```

```python
df["Actor"]        # Single column (as Series)
```

```python
df[["Actor", "Genre"]] 
```

### Selecting Rows by Index
**.loc[]**  (label-based) <br >   **.iloc[]** (position-based):

```python
df.loc[0,"Actor"]   # Value at row 0, column 'Actor'
```

```python
df.iloc[0,1]    # Value at row 0, column at index 1
```

```python
df.loc[0:2, ["Actor", "Genre"]]   # Rows 0 to 2, selected columns
```

```python
df.iloc[0:2, 0:2]                 # Rows and cols by index position
```

### Fast Access: .at and .iat
These are optimized for single element access:

```python
df.at[0, "Actor"]       # Fast label-based access
```

```python
df.iat[0, 1]           # Fast position-based access
```

### Filtering with Conditions
**Simple Condition**

```python
df[df["Year"] > 2020]
```

**Multiple Conditions (AND / OR)**

```python
df[(df["Year"] > 2000) & (df["Genre"] == "Action")]
```

```python
df[(df["Actor"] == "Shah Rukh Khan") | (df["Year"] <2000 )]
```

### Querying with .query()

```python
df.query("Year > 2000 and Genre == 'Action'")
```

```python
col = "Year"
df.query(f"{col} > 2019")
```

### 1. Column names become variables
You can reference column names directly in the query string:

```python
df.query("Year > 2019 and Genre == 'Action'")
```

### 2. String values must be in quotes
Use single or double quotes around strings in the expression:

```python
df.query("Actor == 'Salman Khan'")
```

```python
#If you have quotes inside quotes, mix them:

df.query('Genre == "Horror Comedy"')
```

### 3. Use backticks for column names with spaces or special characters
If a column name has spaces, use backticks (`):

```python
df.query("`Actor` == 'Rajkummar Rao'")
```

### 4. You can use @ to reference Python variables
To pass external variables into .query():

```python
Year_limit = 2019
df.query("Year > @Year_limit")
```

### 5. Logical operators
Use these:

and, or, not — instead of &, |, ~ <br>

==, !=, <, >, <=, >=

```python
df.query("Year > 2019 and Genre == 'Action'")  
```

### 6. Chained comparisons
Just like Python:

```python
df.query("2019 < Year <= 2025")
```

### 7. Avoid using reserved keywords as column names
If you have a column named class, lambda, etc., you’ll need to use backticks:

```python
df.query("`Genre` == 'Action'")
```

### 8. Case-sensitive
Column names and string values are case-sensitive:

```python
df.query("Genre == 'action'")       # ❌ if actual value is 'Action'
```

### 9. .query() returns a copy, not a view
The result is a new DataFrame. Changes won't affect the original unless reassigned:

```python
filtered = df.query("IMDb < 7")
```

```python
print(filtered)
```

```python
filtered_movies = df[df['IMDb'] > 8][['Film', 'IMDb']]
print(filtered_movies)
```
