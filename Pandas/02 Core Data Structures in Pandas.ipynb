# Core Data Structures in Pandas

### Series — 1D Labeled Array
A Series is like a list with labels (index).

```python
import pandas as pd
s = pd.Series([23,56,87,97])
```

```python
s
```

```python
s = pd.Series([23,56,87,97], index= ["gauri","om","parag","mayur"])
```

```python
s
```

### DataFrame — 2D Labeled Table
A DataFrame is like a dictionary of Series — multiple columns with labels.

```python
data = {
    "name": ["gauri", "om", "parag"],
    "age": [75, 98, 65],
    "city": ["yeola", "jalgaon", "satana"]
}

df = pd.DataFrame(data)
print(df)
```

```python
df.index = ["a", "b", "c"]
df.columns = ["Name", "Age", "City"]     
```

```python
df
```
